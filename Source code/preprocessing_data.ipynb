## 5.1 Hàm bổ trợ


```python
def count_nan_location(data, column, columnDepend):
  nan_dict = {}
  location_count = data[column].value_counts()
  mode_location = data.groupby(column)[columnDepend]
  for districts in mode_location.groups:
    total_nan = mode_location.get_group(districts).isnull().sum()
    if total_nan > 0:
      nan_dict[districts] = {'count': total_nan, 'percentage': total_nan / location_count.loc[districts]}
  return nan_dict
```

```python
def get_mode_location(data, column, columnDepend):
  mode_dict = {}
  location_count = data[column].value_counts()
  mode_location = data.groupby(column)[columnDepend]
  for districts in mode_location.groups:
    mode_dict[districts] = mode_location.get_group(districts).mode().values[0]
  return mode_dict
```

```python
def count_nan_room(data, column):
  gv = data.groupby(['building_type','type_squares'])
  return gv[column].apply(lambda x: x.isna().sum())
```

```python
def find_mode_groups(data, groups, column):
  mode_groups_column = {}

  for groups in data.groupby(groups)[column]:
    mode_groups = groups[1].mode()
    if len(mode_groups) > 0:
      mode_groups_column[groups[0]] = mode_groups[0]
  return mode_groups_column
```

```python
def fill_nan_room(row, column, mode_dict):
  if not pd.isnull(row[column]):
    return row
  key = (row['building_type'], row['type_squares'])
  row[column] = mode_dict.get(key, row[column])
  return row
```

Ta có thể có một phương pháp điền khuyết dữ liệu thuộc tính này như sau:



1.   Cố gắng tìm kiếm thêm thông tin dự án từ tên bài đăng và điền tên dự án tương ứng
2.   Sau bước trên, những bất động sản chưa có tên dự án thì sẽ được chuyển thành không dự án để điền khuyết



**Thử điền khuyết tên dự án bằng cách kiếm tìm tên dự án trong tên bài đăng hoặc có tên trong danh sách dự án hiện có**

```python
pattern = []
for name in project_names.index:
  pattern.append({name : name.lower().replace('dự án ', '').split(' ')})
```

```python
pattern
```

```python
name_title = data[data.project_names.isnull()]
```

```python
def is_in_project(project_data, name_check):
  name = str(name_check).lower()
  for p in project_data:
    v = list(p.values())[0]
    is_in = all([s in name for s in v])
    if is_in: return list(p.keys())[0]
  return np.nan
```

```python
results = []

for name in name_title['post_title']:
  name = str(name).lower()
  count = 0
  flag = False
  for p in pattern:
    flag = all([sub_str in name for sub_str in p])
    if flag: break
  results.append(flag)
```

```python
def fill_nan_project(row):
  p = is_in_project(pattern,row['post_title'])
  if pd.isna(row['project_names']) and p != np.nan:
    row['project_names'] = p
  return row
```

```python
data['project_names'] = data.apply(fill_nan_project, axis = 1)['project_names']
```
